```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/countries.csv')  # Update path if necessary

# Display first few rows
df.head()
```

```
# Show basic information about the dataset
df.info()

# Summary statistics for numeric columns
df.describe()
```

```
df.isna().sum()
```

```
# Example conversion (only if needed)
df['Population'] = pd.to_numeric(df['Population'], errors='coerce')
```

```
df.duplicated().sum()
```

```
df.columns = df.columns.str.replace(' ', '_')
```

```
# Central tendency calculations

# Population
pop_mean = df['Population'].mean()
pop_median = df['Population'].median()

# Population Source (Mode)
pop_source_mode = df['Population_source'].mode()[0]

# Quality of Life Index
qol_mean = df['Quality_of_Life_Index'].mean()
qol_median = df['Quality_of_Life_Index'].median()

# Pollution Index
poll_mean = df['Pollution_Index'].mean()
poll_median = df['Pollution_Index'].median()

pop_mean, pop_median, pop_source_mode, qol_mean, qol_median, poll_mean, poll_median
```

```
pop_sd = df['Population'].std()
qol_sd = df['Quality_of_Life_Index'].std()
poll_sd = df['Pollution_Index'].std()

pop_sd, qol_sd, poll_sd
```

```
qol_cv = qol_sd / qol_mean
poll_cv = poll_sd / poll_mean

qol_cv, poll_cv
```

```
# Problem 3: Outlier detection for population density

density = df['Density_pop./km2']

# Calculate quartiles and IQR
Q1 = density.quantile(0.25)
Q3 = density.quantile(0.75)
IQR = Q3 - Q1

lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(density < lower_fence) | (density > upper_fence)]

# Display outlier country names and densities
outlier_countries = outliers[['Country', 'Density_pop./km2']]
outlier_countries
```

```
# Problem 4: Correlation Analysis

# Select relevant columns
corr_cols = [
    'Quality_of_Life_Index',
    'Purchasing_Power_Index',
    'Pollution_Index',
    'Safety_Index',
    'Cost_of_Living_Index',
    'Restaurant_Price_Index',
    'Climate_Index'
]

corr_df = df[corr_cols]

# Compute correlation matrix
corr_matrix = corr_df.corr()

corr_matrix
```

```
plt.figure(figsize=(10, 8))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr_cols)), corr_cols, rotation=45, ha='right')
plt.yticks(range(len(corr_cols)), corr_cols)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```
# Find strongest positive correlation among the variables
corr_matrix_unstacked = corr_matrix.unstack()
strongest_pos_pair = corr_matrix_unstacked.sort_values(ascending=False)

# Display top few correlations
strongest_pos_pair.head(10)
```

```
# Categorise countries into population bands
def pop_band(x):
    if x < 10_000_000:
        return 'Small (<10M)'
    elif x <= 100_000_000:
        return 'Medium (10M100M)'
    else:
        return 'Large (>100M)'

df['Population Band'] = df['Population'].apply(pop_band)

band_counts = df['Population Band'].value_counts()

# Plot bar chart
plt.figure(figsize=(8,6))
plt.bar(band_counts.index, band_counts.values)
plt.xlabel('Population Band')
plt.ylabel('Number of Countries')
plt.title('Countries per Population Band')
plt.show()
```

```
# Count and percentage distribution
source_counts = df['Population_source'].value_counts()
source_pct = round((source_counts / source_counts.sum()) * 100, 2)

# Bar chart of population source percentages
plt.figure(figsize=(8,6))
plt.bar(source_pct.index, source_pct.values)
plt.xlabel('Population Data Source')
plt.ylabel('Percentage of Countries (%)')
plt.title('Percentage Distribution of Population Source')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```
# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(df['Quality_of_Life_Index'], df['Safety_Index'])
plt.xlabel('Quality of Life Index')
plt.ylabel('Safety Index')
plt.title('Correlation: Quality of Life vs Safety Index')
plt.grid(True)
plt.show()

# Display correlation coefficient
qol_safety_corr = df[['Quality_of_Life_Index', 'Safety_Index']].corr().iloc[0,1]
qol_safety_corr
```

```
# Create 10 equal-width bins for Cost of Living Index
min_val = df['Cost_of_Living_Index'].min()
max_val = df['Cost_of_Living_Index'].max()

bins = np.linspace(min_val, max_val, 11)  # 10 bands = 11 boundaries
df['Cost of Living Band'] = pd.cut(df['Cost_of_Living_Index'], bins=bins)

band_distribution = df['Cost of Living Band'].value_counts().sort_index()

# Bar chart of distribution
plt.figure(figsize=(12,6))
plt.bar(band_distribution.index.astype(str), band_distribution.values)
plt.xlabel('Cost of Living Index Bands')
plt.ylabel('Number of Countries')
plt.title('Distribution of Countries Across Cost of Living Index Bands')
plt.xticks(rotation=90)
plt.show()
```
